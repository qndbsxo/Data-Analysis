## 실습 0. 유임승차, 유임하차, 무임승차, 무임하차 4가지 별로,  각각 가장 많은 역을 찾아보세요.

### 유임승차가 가장 많은 역
---

```python
# 유임승차가 가장 많은 역을 기준으로 정렬하기
df.sort_values(by='유임승차', ascending=False)

# 유임승차가 가장 많은 역은 강남역이다
```

### 유임하차가 가장 많은 역
---

```python
# 유임하차 컬럼을 내림차순으로 정렬
df.sort_values(by='유임하차', ascending=False) 

# 유임하차가 가장 많은 역 또한 강남역이다.
```

### 무임승차가 가장 많은 역
---

```python
# 무임승차 컬럼을 기준으로 내림차순으로 정렬
df.sort_values(by='무임승차', ascending=False)

# 무임승차가 가장 많은 역은 종로3가 역이다.
```

### 무임하차가 가장 많은 역
---

```python
# 무임하차 컬럼을 기준으로 내림차순으로 정렬
df.sort_values(by='무임하차', ascending=False)

# 무임하차가 가장 많은 역은 제기동역이다.
```

## 실습 1. 무임승차 대비 유임승차 비율이 가장 높은 역은 어디입니까?

```python
df_free_ride= df[df['무임승차'] > 0] # 무임승차 인원이 0명 이상인 역만 필터링
df['무유임율'] = df_free_ride['유임승차'] / df_free_ride['무임승차'] # 유무임율 컬럼 추가
df.sort_values(by='무유임율', ascending=False)  # 무유임율 컬럼을 기준으로 내림차순 정렬

# 유무임률이 가장 높은 역은 충무로역 입니다
```

## 실습 2. 전체승차인원(유임+무임)이 만명이상인 역 중, 유임승차 비율이 가장 높은 역은 어디입니까?

```python
df['전체인원'] = df['유임승차'] + df['무임승차'] # 전체인원 계산 // 컬럼추가
df_10000 = df[df['전체인원'] >= 10000] # 전체인원이 만 명 이상인역만 filtering
df_10000.sort_values(by='무유임율',ascending=False) # 무임율 컬럼을 기준으로 내림차순 정렬

# 홍대입구역입니다. 
```

## 실습 3. 모든 역의 유임승차, 유임하차, 무임 승차, 무임하차, 총 4개를, 파이차트로 나타내세요.

```python
df_paid_ride = df['유임승차'].sum() # 유임승차 총합
df_paid_quit = df['유임하차'].sum() # 유임하차 총합
df_free_ride = df['무임승차'].sum() # 무임승차 총합
df_free_quit = df['무임하차'].sum() # 무임하차 총합

labels = ['유임승차', '유임하차', '무임승차', '무임하차']
```

```python
# visualization
plt.pie(x=[df_paid_ride, df_paid_quit, df_free_ride, df_free_quit], labels=labels, autopct='%.1f%%')
plt.show()
```
